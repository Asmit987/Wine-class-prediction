```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_wine
```

```python
dataset = load_wine()
dataset
```

```python
data1 = pd.DataFrame(dataset.data, columns = dataset.feature_names)
data1.head()
```

```python
data2 = pd.DataFrame(dataset.target, columns = ['WineClass'])
data2.head()
```

```python
df = pd.concat((data1,data2), axis = 1)
df
```

```python
df.rename(columns={"od280/od315_of_diluted_wines":"diluted_wines"}, inplace=True)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
sns.countplot(df['WineClass'])
```

```python
sns.pairplot(df, hue = 'WineClass')
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
```

```python
y_train_pred = model.predict(X_train)
```

```python
predicted
```

```python
from sklearn import metrics
```

```python
print('Check is the model low bias')
metrics.accuracy_score(y_train_pred, y_train)
```

```python
print('Check is the model low variance')
metrics.accuracy_score(predicted, y_test)
```

```python
print(metrics.confusion_matrix(predicted, y_test))
```

```python
print(metrics.classification_report(predicted, y_test))
```
